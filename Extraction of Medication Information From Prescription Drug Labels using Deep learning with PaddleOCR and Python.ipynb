## 2. Import necessary libraries

```python
from paddleocr import PaddleOCR, draw_ocr  # Importing main OCR dependencies
from matplotlib import pyplot as plt  # Importing library for plotting images
import cv2  # Importing OpenCV library
import os  # Importing library for folder directory navigation
```

## 3. Initializing the OCR model and defining the image path

```python
# Initialize OCR model
ocr_model = PaddleOCR(lang='en')
```

```python
# Define the path to the image
image_path = os.path.join('.', 'Drug_Images', 'drug1.jpg')
```

## 4. Performing OCR on the image

```python
# Perform OCR on the image using the initialized OCR model
result = ocr_model.ocr(image_path)
```

## 5. Processing and displaying the OCR results

```python
# Display the OCR results obtained from processing the image, which contain text and bounding box information
result
```

```python
# Iterate through each OCR result in the 'result' list
for detection in result[0]:
    # Extract and print the recognized text from each result
    print(detection[1][0])
```

```python
# Extracting bounding box coordinates of detected text components
bounding_boxes = [detection[0] for detection in result[0]]

# Extracting recognized text from detected components
recognized_texts = [detection[1][0] for detection in result[0]]

# Extracting confidence scores of detected text components
confidence_scores = [detection[1][1] for detection in result[0]]
```

## 6. Visualizing the results

```python
# Define the file path for the English standard font located in the 'Fonts' directory
font_path = os.path.join('.', 'Fonts', 'en_standard.ttf')
font_path
```

```python
# Import the image using OpenCV
image = cv2.imread(image_path)

# Reorder the color channels from BGR to RGB
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
```

```python
# Set the size of the display area for visualization
plt.figure(figsize=(16, 16))

# Draw annotations (bounding boxes and text) on the image
annotated_image = draw_ocr(image, bounding_boxes, recognized_texts, confidence_scores, font_path=font_path)

# Display the annotated image using matplotlib
plt.imshow(annotated_image)

# Save the annotated image
plt.savefig('annotated_image.jpg')
```
